```python
import tensorflow as tf
import pandas as pd 
import os 
import numpy as np 
```

```python
zip_path = tf.keras.utils.get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)
csv_path, _ = os.path.splitext(zip_path)
```

```python
df = pd.read_csv(csv_path)
df
```

```python
df = df[5::6]
df
```

```python
df.index = pd.to_datetime(df['Date Time'], format = '%d.%m.%Y %H:%M:%S')
df
```

```python
temp = df['T (degC)']
temp.plot()
```

```python
df_temp = pd.DataFrame({'temprature' : temp})
df_temp['second'] = df_temp.index.map(pd.Timestamp.timestamp)
df_temp
```

```python
sec_day = 60*60*24
sec_year = sec_day * 365.2425

df_temp['day_sin'] = np.sin(df_temp['second'] * 2*np.pi/sec_day)
df_temp['day_cos'] = np.cos(df_temp['second'] * 2* np.pi/sec_day)
df_temp['year_sin'] = np.sin(df_temp['second'] * 2*np.pi/sec_year)
df_temp['year_cos'] = np.cos(df_temp['second'] * 2*np.pi/sec_year)
```

```python
df_temp = df_temp.drop('second',axis = 1)
df_temp.head()
```

```python
df_temp_pres = pd.concat([df['p (mbar)'],df_temp],axis = 1)
df_temp_pres.head()
```

```python
def df_x_y(df,window_size):
    df_np = df.to_numpy()
    x = []
    y = []
    for i in range(len(df_np) - window_size):
        row = [r for r in df_np[i: i+ window_size]]
        x.append(row)
        labels = df_np[i + window_size][0],df_np[i + window_size][1]
        y.append(labels)
    return np.array(x),np.array(y)
```

```python
x2,y2 = df_x_y(df_temp_pres ,7)
x2.shape,y2.shape
```

```python
x_train3,y_train3 = x2[:60000] , y2[: 60000]
x_val3,y_val3 = x2[60000: 65000] , y2[60000:65000]
x_test3 ,y_test3 = x2[65000:],y2[65000:]
```

```python
x_train3.shape,y_train3.shape,x_val3.shape,y_val3.shape,x_test3.shape,y_test3.shape
```

```python
p_mean = np.mean(x_train3[:,:,0])
p_std = np.std(x_train3[:,:,0])

temp_mean = np.mean(x_train3[:,:,1])
temp_std = np.mean(x_train3[:,:,1])
```

```python
def scaling_val_in(df):
    df[:,:,0] = (df[:,:,0] - p_mean) / p_std
    df[:,:,1] = (df[:,:,1] - temp_mean) / temp_std
    return df

def scaling_val_out(df):
    df[:,0] = (df[:,0] - p_mean) / p_std
    df[:,1] = (df[:,1] - temp_mean) / temp_std
    return df
```

```python
scaling_val_in(x_train3)
scaling_val_in(x_test3)
scaling_val_in(x_val3)

scaling_val_out(y_train3)
scaling_val_out(y_val3)
scaling_val_out(y_test3)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
```

```python
model1 = Sequential()
model1.add(InputLayer((7,6)))
model1.add(LSTM(128))
model1.add(Dense(8,'relu'))
model1.add(Dense(2,'linear'))

model1.summary()
```

```python
cp5 = ModelCheckpoint('model1/best_model.keras',save_best_only = True)
model1.compile(loss = MeanSquaredError(),optimizer = Adam(learning_rate = 0.0001) , metrics = [MeanSquaredError()])
```

```python
model1.fit(x_train3, y_train3,validation_data =  (x_val3,y_val3),epochs = 10,callbacks = [cp5])
```

```python
def plot_predictions1(model, X, y, start=0, end=100):
  predictions = model.predict(X)
  p_preds, temp_preds = predictions[:, 0], predictions[:, 1]
  p_actuals, temp_actuals = y[:, 0], y[:, 1]
  df = pd.DataFrame(data={'Temperature Predictions': temp_preds,
                          'Temperature Actuals':temp_actuals,
                          'Pressure Predictions': p_preds,
                          'Pressure Actuals': p_actuals
                          })
  plt.plot(df['Temperature Predictions'][start:end])
  plt.plot(df['Temperature Actuals'][start:end])
  plt.plot(df['Pressure Predictions'][start:end])
  plt.plot(df['Pressure Actuals'][start:end])
  return df[start:end]
    
```

```python
plot_predictions1(model1,x_test3,y_test3,0,2000)
```

```python
def after_scaling_p(nums):
    nums = (nums * p_std) + p_mean
    return nums

def after_scaling_temp(nums):
    nums  = (nums* temp_std) + temp_mean
    return nums
```

```python
import matplotlib.pyplot as plt
def plot_predictions2(model, X, y, start=0, end=100):
  predictions = model.predict(X)
  p_preds, temp_preds = after_scaling_p(predictions[:, 0]), after_scaling_temp(predictions[:, 1])
  p_actuals, temp_actuals = after_scaling_p(y[:, 0]), after_scaling_temp(y[:, 1])
  df = pd.DataFrame(data={'Temperature Predictions': temp_preds,
                          'Temperature Actuals':temp_actuals,
                          'Pressure Predictions': p_preds,
                          'Pressure Actuals': p_actuals
                          })
  plt.plot(df['Temperature Predictions'][start:end])
  plt.plot(df['Temperature Actuals'][start:end])
  plt.plot(df['Pressure Predictions'][start:end])
  plt.plot(df['Pressure Actuals'][start:end])
  return df[start:end]
    
```

```python
after_process = plot_predictions2(model1,x_test3,y_test3,0,2000)
after_process
```

```python
start,end = 0, 1000
plt.plot(after_process['Temperature Predictions'][start:end])
plt.plot(after_process['Temperature Actuals'][start:end])
```
